```python
import pandas as pd
from algormeter import *
from algormeter.libs import *
import algormeter.algorithms as alg
```

```python
problems = [ # number are X dimensions
            (Parab,[2,4,6]),
            (ParAbs,[2]),
            (Acad,[2]),
            (JB01,[2]), 
        ]

def gradient(p, **kwargs):
    for k in p.loop():
        p.Xkp1 = p.Xk - 1/(k+1) * p.gfXk / np.linalg.norm(p.gfXk) 

algorithms = [
                gradient,
                alg.gradient,
        ]

df, pv = algorMeter(algorithms = algorithms, problems = problems, iterations = 200 )
display (pv)
display(df)
```
